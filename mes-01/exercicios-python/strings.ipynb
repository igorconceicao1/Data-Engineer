STRINGS PROPERTIES AND METHODS

Immutability -  Are strings mutable?

Strings are not mutable! (meaning you can't use indexing to change individual elements of a string)

```python
name = "Sam"
print(name[::])

# name[0] = "P"

last_letters = name[1:]
print(last_letters)

print("P" + last_letters)

```

```python
name2 = "Igor"
print(name2[::])
#Ogoro

mid_name = name2[1:]
print(mid_name)

print('O' + mid_name + 'o')

mid_name = name2[1:] + 'teste'
print(mid_name)

letter = 'z'
letter * 10
```

```python
x = 'Hello World'
print(x.upper()) #tudo mainsculo
print(x) #nao ativa na variavel principal
print(x.lower()) #minusculo
print(x.split()) #cria uma lista da string
```

STRING INTERPOLATION

.format

```python
print('This is a string {}'.format('INSERTED'))
```

```python
print('The {0} {0} {0}'.format('fox','brown','quick'))
```

```python
print('The {q} {b} {f}'.format(f='fox', b='brown', q='quick'))
```

```python
print("Hello {r}, my name is {i}".format(r='test', i='Igor'))
```

{value:width.precision f} - FLOAT FORMATTING FOLLOW

```python
result = 100/7777
print(result)

print("The result was {r:1.11f}".format(r=result))

nome = "Josse"
print("Hello, his name is {n}".format(n = nome))
```

f-strings - FORMATTED STRING LITERALS

```python
name = "Josse"
print(f'Hello, his name is {name}')

print('{p} {r}'.format(p='Python', r='rules!'))

p = 'Python'
r = 'rules!'
print (f'{p} {r}')


x = 'Carlos'
b = 'Jonatan'
print('Pessoa 1: {p1} | Pessoa 2: {p2}'.format(p1=x, p2=b))
```
