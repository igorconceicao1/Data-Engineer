```python
resultado = round(0.1 + 0.2 - 0.53453654
, 5)  # Arredonda para 5 casas decimais
print(resultado)  # Saída: 0.0
```

```python
#Usar o módulo decimal
from decimal import Decimal
resultado = Decimal('0.1') + Decimal('0.2') - Decimal('0.3')
print(resultado)  # Saída: 0.0
```

```python
#Alternativa com math.isclose
import math
resultado = 0.1 + 0.2 - 0.3
print(math.isclose(resultado, 0.0, abs_tol=1e-10))  # Saída: True
```

```python
#ponto flutuante
valor1 = 10  # 0.1 em centavos
valor2 = 20  # 0.2 em centavos
valor3 = 30  # 0.3 em centavos
resultado = (valor1 + valor2 - valor3) / 100  # Converte de volta
print(resultado)  # Saída: 0.0
```
