```python
my_dic = {'key1':'value1', 'key2':'value2'}
print(my_dic)

my_dic['key1']

feira = {'apple':3, 'banana':0.90, 'pear':2.50, 'sucos':{
    'tipos':['laranja', 'uva', 'maca']}}

feira['sucos']['tipos'][0].upper() #eu criei um dic suco e uma lista com os tipos de suco dentro dessa key
```

```python
feira['lemon'] = 5.55
print(feira)
```

```python
#to see all the keys from a dictionary
feira.keys()
```

```python
#to see all the values from a dictionary
feira.values()
```
