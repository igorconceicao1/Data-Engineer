.append() and .pop(), adiciona no final da lista e elemina da lista

```python
new_list.append('66') #adciona no final da lista
print(new_list)
popped_item = new_list.pop(0)#elimina da lista
print(new_list)
print(popped_item)
```

.sort() - organize the list

```python
new_list2 = ['a','e','x','t','f','b']
print(new_list2)
new_list2.sort()
print(new_list2)
my_sorted_list = new_list2 #()com parenteses ou sort aqui, retorna None
print(my_sorted_list)
```

.reverse()

```python
my_sorted_list.reverse()
print(my_sorted_list)
```

```python
my_list4 = [1, 1, [1, 2]]
print(f"Lista interna: {my_list4[2]}")    # Saída: [1, 2]
print(f"Elemento desejado: {my_list4[2][1]}")  # Saída: 2

# Exemplo mais complexo
nested_list = [1, [2, [3, 4]]]
print(f"Mais aninhado: {nested_list[1][1][1]}")  # Saída: 4
```
