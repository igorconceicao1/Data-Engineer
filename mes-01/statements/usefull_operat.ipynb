```python
#enumerate
index_count = 0

for letter in 'abcde':
    print(f'At index {index_count} the letter is {letter}')
    #print('At index {} the letter is {}'.format(index_count, letter))
    index_count += 1
```

```python
#por ser uma forma muito comum que vamos usar futuramente, por isso o python criou a enumerate function
index_count = 0
world = 'abcde'

for letter in world:
    print(world[index_count])
    index_count += 1
```

```python
#enumerate // me retorna tuplas
world = 'abcdef'

for item in enumerate(world):
    print(item)
```

```python
world = 'abcdef'

for index, item in enumerate(world):
    print(index)
    print(item)
    print('\n')
```

```python
#zip function // retorna uma tupla tambem, be careful, zip togueder as far as the list which is the 
#shortest. Traducao: o zip so vai funcionar se as listas tiverem os mesmos tamanhos
#MESMO SE TIVER MAIS NUMEROS EM UMA LISTA SO VAI RETORNAR A QUANTIDADE EXATA NAS DUAS LISTAS

list1 = [1,2,3,4,5,6]
list2 = ['a', 'b', 'c', 'd', 'e', 'f']

for item in zip(list1, list2):
    print(item)

list(zip(list1, list2))
```

```python
#se eu quero unpacking

list1 = [1,2,3,4,5,6]
list2 = ['a', 'b', 'c', 'd', 'e', 'f']

for a,b in zip(list1, list2):
    print(a)
    print(b)
    print('\n')
```
