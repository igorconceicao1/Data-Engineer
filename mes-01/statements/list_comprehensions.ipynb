List Comprehensions are a unique way of quickly creating a list with python
if you are using a for loop along with .append() to createe a list; List Comprehensions are good alternative 

```python
mystring = 'hello'
my_list = []

for letter in mystring:
    my_list.append(letter)

my_list
```

more efficient way do this with python:
METHOD List Comprehensions
just take that element for element in some other iterable object such as another list

```python
mystring = 'hello world'
mylist = [letter for letter in mystring]

mylist
```

```python
mylist = [letter for letter in 'woldtest']

mylist
```

we gonna see a lot this

```python
mylist = [num for num in range(0,11)]

mylist
```

begin to perform operations in the first element and complex arithmetic

```python
mylist = [num**2 for num in range(0,11)]

mylist
```

```python
celcius = [0, 10, 20, 34.5, 77]
fahrenheit = [((9/5)*temp + 32) for temp in celcius]

fahrenheit
```

```python
celcius = [0, 10, 20, 34.5, 77]
fahrenheit = []

for temp in celcius:
    temp = (9/5)*temp + 32
    fahrenheit.append(temp)
fahrenheit
```

```python
celcius = [0, 10, 20, 34.5, 77]
fahrenheit = []

for temp in celcius:
    fahrenheit.append(((9/5)*temp + 32))
fahrenheit
```

you can also - add in IF statements into this

```python
mylist = [num for num in range(0,11) if num%2==0]

mylist
```

you also can do - IF and ELSE statments - but READABILITY AND REPRODUCIBILITY in the Python code should be the most important thing.
List Comprehensions can kind of fool you into a thinking youre becoming a really good programmer when in fact youre writing just kind of really one-liners that are really ugly and its hard to read 

```python
results = [num if num%2==0 else 'ODD' for num in range(0,11)]

results
```

you can alsdo do - nested loops 

```python
 #nested loops
mylist2 = []

for x in [2,4,6]:
    for y in [200,400,600]:
        mylist2.append(x*y)

print(mylist2)
```

```python
mylist2 = [x*y for x in [2,4,6] for y in [200,400,600]]

mylist2
```
