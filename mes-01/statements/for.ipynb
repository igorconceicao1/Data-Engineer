```python
#tupla unpacking - where you kind of duplicate the structure of the items
mylist = [(1,2), (3,4), (5,6), (7,8)]
len(mylist)

for i in mylist:
    print(i)

for a,b in mylist:
    print(a)
    print(b)
```

```python
#podemos usar a regra de tupla unpacking com dictionaries tambem
d = {'k1':1, 'k2':2, 'k3':3}

for key,value in d.items():
    print(value)
```

```python
#dictionary
d = {'k1':1, 'k2':2, 'k3':3}

for i in d:
    print(i)
```

```python
d = {'k1':1, 'k2':2, 'k3':3}

for i in d.items():
    print(i)
```

```python
d = {'k1':1, 'k2':5, 'k3':3}

for key in d.keys():
    print(key)

for value in d.values():
    print(value)
```
