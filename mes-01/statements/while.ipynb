```python
# break - breaks out of the current closet enclosing loop
# continue - goes to the top of the closest enclosing loop
# pass - does nothing at all
```

```python
mystring = 'Sammm'

for i in mystring:
    if i == 'a': #if the letter is a, go ahead and continue and go back to the loop
        continue
    print(i)
```

```python
mystring = 'Sammm'

for i in mystring:
    if i == 'a': #just stops this loop if the letter is equal to 'a'
        break
    print(i)
```
